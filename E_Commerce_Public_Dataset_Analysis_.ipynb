### Cleaning Data

```python
# mengatasi missing value pada data orders

# Menghitung rata-rata waktu dari purchase ke approval
average_approval_time = (orders_df['order_approved_at'] - orders_df['order_purchase_timestamp']).mean()
# Menghitung rata-rata waktu dari approval ke carrier
average_carrier_time = (orders_df['order_delivered_carrier_date'] - orders_df['order_approved_at']).mean()
# Menghitung rata-rata waktu dari carrier ke customer
average_delivery_time = (orders_df['order_delivered_customer_date'] - orders_df['order_delivered_carrier_date']).mean()

# Mengisi missing value pada order_approved_at
orders_df['order_approved_at'] = orders_df['order_approved_at'].fillna(orders_df['order_purchase_timestamp'] + average_approval_time)

# Mengisi missing value pada order_delivered_carrier_date
orders_df['order_delivered_carrier_date'] = orders_df['order_delivered_carrier_date'].fillna(orders_df['order_approved_at'] + average_carrier_time)

# Mengisi missing value pada order_delivered_customer_date
orders_df['order_delivered_customer_date'] = orders_df['order_delivered_customer_date'].fillna(orders_df['order_delivered_carrier_date'] + average_delivery_time)

# Menampilkan data setelah pengisian missing value
orders_df.isna().sum()
```

```python
# mengatasi missing value pada products

# 1. Mengisi nilai null dengan 'N/A' pada kolom product_category_name
products_df['product_category_name'] = products_df['product_category_name'].fillna('N/A')

# 2. Mengisi nilai null dengan 0 pada kolom product_name_lenght, product_description_lenght, dan product_photos_qty
products_df['product_name_lenght'] = products_df['product_name_lenght'].fillna(0)
products_df['product_description_lenght'] = products_df['product_description_lenght'].fillna(0)
products_df['product_photos_qty'] = products_df['product_photos_qty'].fillna(0)

# 3. Mengisi nilai 0 dengan null pada kolom product_weight_g, product_width_cm, product_height_cm, dan product_length_cm
products_df['product_weight_g'] = products_df['product_weight_g'].replace(0, np.nan)
products_df['product_width_cm'] = products_df['product_width_cm'].replace(0, np.nan)
products_df['product_height_cm'] = products_df['product_height_cm'].replace(0, np.nan)
products_df['product_length_cm'] = products_df['product_length_cm'].replace(0, np.nan)

products_df.isna().sum()
```

## Exploratory Data Analysis (EDA)

### Produk apa yang paling banyak terjual?

```python
# Menggabungkan order_items dengan products
order_items_products_df = pd.merge(order_items_df, products_df, on='product_id', how='left')

# Menggabungkan orders dengan order_items_products
orders_order_items_products_df = pd.merge(orders_df, order_items_products_df, on='order_id', how='left')

# Menggabungkan orders_order_items_products dengan order_payments
full_df = pd.merge(orders_order_items_products_df, order_payments_df, on='order_id', how='left')

# Konversi order_purchase_timestamp menjadi datetime
full_df['order_purchase_timestamp'] = pd.to_datetime(full_df['order_purchase_timestamp'], errors='coerce')
```

```python
# Melihat produk yang paling laris terjual
top_products = full_df['product_id'].value_counts().head(10)
top_products = top_products.rename_axis('product_id').reset_index(name='count')

#Menggabungkan dengan products_df untuk mendapatkan nama kategori produk
top_products = pd.merge(top_products, products_df[['product_id', 'product_category_name']], on='product_id', how='left')

# Menampilkan produk yang paling laris terjual beserta nama kategori produk
print("Top 10 produk yang paling laris terjual:")
print(top_products)
```

```python
# Menggabungkan dengan product_category untuk mendapatkan nama kategori produk dalam bahasa Inggris
top_products = pd.merge(top_products, product_category_df, on='product_category_name', how='left')
top_products.rename(columns={'product_category_name': 'product_category_name_indonesian', 'product_category_name_english': 'product_category_name'}, inplace=True)

# Menampilkan produk yang paling laris terjual beserta nama kategori produk dalam bahasa inggris
print("Top 10 produk yang paling laris terjual:")
print(top_products[['product_category_name']])
```

```python
# Menambah kolom tahun
full_df['year'] = full_df['order_purchase_timestamp'].dt.year

# Mengidentifikasi produk terlaris untuk setiap tahun
top_products_per_year = full_df.groupby(['year', 'product_id']).size().reset_index(name='count')
top_products_per_year = top_products_per_year.sort_values(['year', 'count'], ascending=[True, False])

# Mendapatkan produk terlaris untuk setiap tahun
top_products_2016 = top_products_per_year[top_products_per_year['year'] == 2016].head(1)
top_products_2017 = top_products_per_year[top_products_per_year['year'] == 2017].head(1)
top_products_2018 = top_products_per_year[top_products_per_year['year'] == 2018].head(1)

# Menampilkan produk terlaris untuk setiap tahun
print("Produk terlaris tahun 2016:")
print(top_products_2016)
print("\nProduk terlaris tahun 2017:")
print(top_products_2017)
print("\nProduk terlaris tahun 2018:")
print(top_products_2018)

# Menggabungkan produk terlaris untuk setiap tahun ke dalam satu tabel
top_products_all_years = pd.concat([top_products_2016, top_products_2017, top_products_2018])
```

### Berapa total pendapatan yang dihasilkan selama bulan september 2016 hingga september 2018?

```python
# Menggabungkan orders dengan order_items
orders_order_items_df = pd.merge(orders_df, order_items_df, on='order_id', how='inner')

# Menggabungkan hasil dengan order_payments
orders_full_df = pd.merge(orders_order_items_df, order_payments_df, on='order_id', how='inner')

# Mengonversi kolom 'payment_value' menjadi tipe data numerik, dengan kesalahan yang tidak dapat dikonversi diubah menjadi NaN
orders_full_df['payment_value'] = pd.to_numeric(orders_full_df['payment_value'], errors='coerce')

# Menghapus baris dengan nilai NaN pada kolom 'payment_value'
orders_full_df = orders_full_df.dropna(subset=['payment_value'])

# Mengonversi kolom 'order_purchase_timestamp' menjadi datetime
orders_full_df['order_purchase_timestamp'] = pd.to_datetime(orders_full_df['order_purchase_timestamp'])

# Membuat kolom tahun dan bulan
orders_full_df['year'] = orders_full_df['order_purchase_timestamp'].dt.year
orders_full_df['month'] = orders_full_df['order_purchase_timestamp'].dt.month
orders_full_df['year_month'] = orders_full_df['order_purchase_timestamp'].dt.to_period('M')

# Menghitung total pendapatan berdasarkan tahun dan bulan
monthly_revenue = orders_full_df.groupby('year_month')['payment_value'].sum().reset_index()

# Menghitung total pendapatan keseluruhan
total_revenue = orders_full_df['payment_value'].sum()

print("Total Pendapatan yang Dihasilkan:", total_revenue)
```

```python
# Gabungkan semua tabel ke dalam satu DataFrame tunggal
all_df = pd.merge(customers_df, orders_df, on='customer_id', how='outer')
all_df = pd.merge(all_df, order_items_df, on='order_id', how='outer')
all_df = pd.merge(all_df, order_payments_df, on='order_id', how='outer')
all_df = pd.merge(all_df, products_df, on='product_id', how='outer')
all_df = pd.merge(all_df, sellers_df, on='seller_id', how='outer')
all_df = pd.merge(all_df, product_category_df, on='product_category_name', how='outer')

# Simpan dalam all_df
all_df.to_csv('all_data.csv', index=False)  # Simpan sebagai CSV
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk apa yang paling banyak terjual?

```python

# Konversi order_purchase_timestamp menjadi datetime
full_df['order_purchase_timestamp'] = pd.to_datetime(full_df['order_purchase_timestamp'], errors='coerce')

# Menghitung jumlah produk yang terjual untuk setiap kategori di setiap tahun
category_sales_per_year = full_df.groupby(['year', 'product_category_name']).size().reset_index(name='count')

# Mengurutkan kategori berdasarkan jumlah produk yang terjual secara menurun untuk setiap tahun
category_sales_per_year_sorted = category_sales_per_year.sort_values(['year', 'count'], ascending=[True, False])

# Memisahkan data untuk tiap tahun
sales_2016 = category_sales_per_year_sorted[category_sales_per_year_sorted['year'] == 2016]
sales_2017 = category_sales_per_year_sorted[category_sales_per_year_sorted['year'] == 2017]
sales_2018 = category_sales_per_year_sorted[category_sales_per_year_sorted['year'] == 2018]

# Plot untuk tahun 2016
plt.figure(figsize=(10, 6))
plt.barh(sales_2016['product_category_name'], sales_2016['count'], color='blue')
plt.xlabel('Jumlah Produk Terjual')
plt.ylabel('Kategori Produk')
plt.title('Penjualan Produk Terlaris per Kategori Tahun 2016')
plt.gca().invert_yaxis()  # Memutar sumbu Y agar kategori teratas berada di bagian atas
plt.tight_layout()
plt.show()

# Plot untuk tahun 2017
plt.figure(figsize=(10, 6))
bars = plt.barh(sales_2017['product_category_name'], sales_2017['count'], color='green')
plt.xlabel('Jumlah Produk Terjual')
plt.ylabel('Kategori Produk')
plt.title('Penjualan Produk Terlaris per Kategori Tahun 2017')
plt.gca().invert_yaxis()  # Memutar sumbu Y agar kategori teratas berada di bagian atas

# Mengatur rotasi dan penempatan label sumbu Y agar tidak bertabrakan
plt.yticks(rotation=0, ha='right', fontsize=5)
plt.tight_layout()
plt.show()

# Plot untuk tahun 2018
plt.figure(figsize=(10, 6))
bars = plt.barh(sales_2018['product_category_name'], sales_2018['count'], color='orange')
plt.xlabel('Jumlah Produk Terjual')
plt.ylabel('Kategori Produk')
plt.title('Penjualan Produk Terlaris per Kategori Tahun 2018')
plt.gca().invert_yaxis()  # Memutar sumbu Y agar kategori teratas berada di bagian atas

# Mengatur rotasi dan penempatan label sumbu Y agar tidak bertabrakan
plt.yticks(rotation=0, ha='right', fontsize=5)
plt.tight_layout()
plt.show()
```
